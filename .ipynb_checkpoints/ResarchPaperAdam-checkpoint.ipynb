## Predictive model 

Here I would like use two machine learning models, with the sckikit-sklearn library, to try to predict whether a client will default on his/her next payment. For this, I will be using the full dataset.

### Load Data and Data Preprocessing

```python
import pandas as pd
```

```python
path = "/Users/fadam/Downloads/"
df = pd.read_csv(path+"credit_card_default.csv")
df.head(5)
```

```python
df.columns
```

### Data Preprocessing
We first need to make sure the data is properly formatted and normalized to build our predictive model. 
   
- Assign target (y) and explanatory variables (X)
- Solve imbalance for target values
- Normalizing the numeric variables
- Split data set into test and train

```python
X = df.drop(['ID', 'default payment_next_month'], axis=1)
y = df['default payment_next_month']
```

```python
y.value_counts()
```

There is a clear imbalance between default and non-default values. We can use SMOTE from imblearn as an oversampling method to balance the distribution.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm , y_sm = smote.fit_resample(X,y)
```

```python
y_sm.value_counts()
```

We can use StandardScalar from sklearn to standardize our numerical variables

```python
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X_sm)
```

We will split data into 80% train and 20% test set.

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_scaled,y_sm,test_size=.2,random_state=42,stratify=y_sm)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
pred = log_reg.predict(X_test)
score = log_reg.score(X_test, y_test)
print(f"Accuracy: {score:.2%}")
```

### K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
pred_y = knn.predict(X_test)
score = knn.score(X_test, y_test)
print(f"Accuracy: {score:.2%}")
```

The KNN model seems to be more accurate than the Logistic Regression, so let's fine tune neighbors parameter in range 1 through 100 with step 5 and run it again. We can use GridSearchCV from sklearn for this.

```python
k_vals = {}
for i in range(1,100,5):
    classifier = KNeighborsClassifier(n_neighbors=i)
    classifier.fit(X_train,y_train)
    k_vals['k='+str(i)] = classifier.score(X_test, y_test)
```

```python
k_vals
```

Let's rerun KNN with the optimal k value and see what accuracy we can get

```python
knn = KNeighborsClassifier(n_neighbors=1, algorithm='kd_tree')
knn.fit(X_train,y_train)
pred_y = knn.predict(X_test)
score = knn.score(X_test, y_test)
print(f"Accuracy: {score:.2%}")
```
